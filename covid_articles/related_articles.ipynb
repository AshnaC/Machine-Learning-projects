```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import scispacy
import spacy
import en_core_sci_lg

# Progress bar implementation
from tqdm import tqdm

# python Seralizer
import joblib

from wordcloud import WordCloud

from scipy.spatial.distance import jensenshannon

from sklearn.decomposition import LatentDirichletAllocation

from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Processed text from clean data
df = pd.read_csv('covid_df.csv')
df.head(2)
```

```python
df_filtered = df[df['body_text'].apply(len) >300]
```

```python
body_text = df['body_text'][df['body_text'].apply(len) >300]
```

```python
# we are using en_core_sci_lg for clean,lemmitize texts
# en_core_sci_lg - A full spaCy pipeline for biomedical data with a larger 
# vocabulary and 600k word vectors.
formatter = en_core_sci_lg.load(disable=["tagger", "parser", "ner"])

# Added to eliminate this error.
#Text of length 1002256 exceeds maximum of 1000000. The v2.x parser
# and NER models require roughly 1GB of temporary memory per 100,000 characters in the input.
# If you're not using the parser or NER, it's probably safe to increase the `nlp.max_length` limit
formatter.max_length =6000000
```

```python
# The stop words are determined by running algo with all words
# Manually checking max occurance words from word_count_df and eliminate meaningless words
# Lda is run on the unfiltered data set
# We can see topics forming with gibberish words so eliminating them as well
specific_stop_words = ['et','al.','fig.','figure', 'de', 'la','fig','al'
                       'http://en.wikipedia.org/wiki/','http://en.wikipedia.org/wiki/business_continuity_planning',
                      'http://en.wikipedia.org/wiki/business_continuity_planning',
                       r'\usepackage{amsbsy', r'\usepackage{amsfonts', r'\usepackage{mathrsfs',
                       r'\usepackage{amssymb', r'\usepackage{wasysym',r'\setlength{\oddsidemargin}{-69pt', 
                       r'\usepackage{upgreek', r'\documentclass[12pt]{minimal',
                      'preprint',  'copyright' ,'holder' 'medrxiv', 'author/funder', 'peer-reviewed',
                       'doi','international' ,'biorxiv', 'cc-by-nc-nd','q' ,'preprint',
                       '.', 'https://doi.org/10.1101/2020.04.01.20050229' ,'doi:', 'medRxiv' ,'preprint' ]
# Mark them as stop words
for w in specific_stop_words:
    formatter.vocab[w].is_stop = True
```

```python
def tokenizer(sentence):
    word_count =[]
    for word in formatter(sentence):
        if not(word.like_num or word.is_stop or word.is_punct or word.is_space or len(word)==1):
            word_count.append(word.lemma_)  
    return word_count
```

```python
from os.path import isfile
if isfile('vectorizer.csv') & isfile('text_vector.csv'):
    vectorizer = joblib.load('vectorizer.csv')
    text_vector = joblib.load('text_vector.csv')
else:
    vectorizer = CountVectorizer(tokenizer=tokenizer,min_df=5)
    # sparse matrix
    text_vector = vectorizer.fit_transform(tqdm(body_text))
    joblib.dump(vectorizer, 'vectorizer.csv')
    joblib.dump(text_vector, 'text_vector.csv')
```

```python
# Applying LDA
if isfile('lda.csv'):
    lda = joblib.load('lda.csv')
else:
    lda = LatentDirichletAllocation(n_components=50,random_state=0)
    lda.fit(text_vector)
    joblib.dump(lda,'lda.csv')
```

```python
if isfile('lda_topic_df.csv'):
    lda_topic_df = pd.read_csv('lda_topic_df.csv')
else:
    lda_topic_df = pd.DataFrame(lda.transform(text_vector))
    lda_topic_df.to_csv('lda_topic_df.csv', index=False)
```

```python
def getNeighbourArticles(paper_id, n_neighbours = 5, plotSim = False):
    paper_idx = df_filtered[df_filtered['sha']== paper_id].index
    
    if len(paper_idx) < 1:
        return 'No paper with id'
    
    loc = df_filtered.index.get_loc(paper_idx[0])
    
    distribution_mat = lda_topic_df.iloc[loc]
    
    distances = lda_topic_df.apply(lambda art: jensenshannon(art,distribution_mat),axis=1)
    
    sorted_idx =distances.argsort()[0:n_neighbours+1]
    #print('sorted_idx',sorted_idx)
    nn_articles = df_filtered.iloc[sorted_idx].reset_index(drop=True)
    
    nn_articles['distances'] = distances[sorted_idx].reset_index(drop=True)
    
    if(plotSim):
        recc_df = lda_topic_df.iloc[sorted_idx.values].transpose()
        selectedIndex = sorted_idx.values[0]
        for i in sorted_idx.values[1:]:
            recc_df[[selectedIndex,i]].plot(kind="bar",figsize=(20,6))
            
    return nn_articles
```

```python
def getRelatedArticles(text , n_neighbours = 5, plotSim = False):
    
    count_vect = vectorizer.transform(text)
    
    distribution_mat = lda.transform(count_vect)[0]
    
    distances = lda_topic_df.apply(lambda art: jensenshannon(art,distribution_mat),axis=1)
    
    sorted_idx =distances.argsort()[0:n_neighbours]
    
    nn_articles = df_filtered.iloc[sorted_idx].reset_index(drop=True)

    nn_articles['distances'] = distances[sorted_idx].reset_index(drop=True)
    
    if(plotSim):
        recc_df = lda_topic_df.iloc[sorted_idx.values]
        
        selectedIndex = 'Beggle'
       
        recc_df =recc_df.append(pd.Series(distribution_mat, name = selectedIndex))
        
        recc_df = recc_df.transpose()
        
        #print(recc_df)
        
        for i in sorted_idx.values:
            recc_df[[selectedIndex,i]].plot(kind="bar",figsize=(20,6))
            
    return nn_articles
```

```python
text= ["Range of incubation periods for the disease in humans (and how this varies across age and health status) and how long individuals are contagious, even after recovery."]
nn_articles = getRelatedArticles(text,  5, True)
```
